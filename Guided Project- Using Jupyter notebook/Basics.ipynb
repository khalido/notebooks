## How does length of employee titles correlate to salary?

steps:

- compare salary to length - how? perhaps divide salary by title length? and add a column?

```python
white_house['Title Length'] = white_house['Position Title'].apply(lambda x: len(x))
white_house['Salary_Title'] = white_house['Salary'] / white_house['Title Length']
white_house
```

```python
import seaborn as sns
#plotting the salary vs title length
sns.lmplot(data=white_house, y ='Salary', x='Title Length')
```
