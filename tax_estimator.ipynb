```python
import matplotlib.pyplot as plt
import numpy as np
```

# Simple revenue estimator

```python
# assumptions

tax_rate = 0.29  # from FBR
income_tax = 6943344  # Actual tax paid

decent_revenue = 10 ** 6

# using a range of margins
margins = np.arange(0.1, 0.4, 0.05)
```

## Calculating revenue

```python
daily_revenue = []
profit = income_tax / tax_rate
print(f"{profit=:,.0f}")

for margin in margins:

    revenue = profit / margin

    daily = revenue / 365  # open 7 days a week
    print(f"{margin=:.1f} {revenue=:,.0f} {daily=:,.0f}")
    daily_revenue.append(daily)
```
